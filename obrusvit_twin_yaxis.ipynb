# Plot with two independent y-axis

Let's see how to plot two graphs into one plot but with two corresponding Y-Axis on each side.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y1 = np.linspace(0,10, 100)
y2 = np.linspace(0,100,100)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.title("This is NOT what we want");
```

## Let's use the object oriented approach of PyPlot.

```python
fig, ax = plt.subplots()
ax.set_title("This is what we want")

# first plot
ax.plot(x, y1, color='b')
ax.set_ylabel("y1", color='b')
ax.tick_params('y', colors='b')
ax.set_ylim(0,10)

# second plot
ax2 = ax.twinx()         # creates second Axes object with invisible x-Axis but independent y-Axis

ax2.plot(x, y2, color='r')
ax2.set_ylabel("y2", color='r')
ax2.tick_params('y', colors='r')
ax2.set_ylim(0,200)

# label of x-axis
ax.set_xlabel("x")
ax.set_xlim(0,10);
```
